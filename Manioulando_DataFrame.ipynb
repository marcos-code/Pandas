```python
import pandas as pd
```

```python
# Criando Dicionarios
dados = {
    'nome': ['Marcos', 'Miguel', 'Angelina'],
    'idade': [29, 59, 60],
    'cidade':['São Paulo', 'Bahia', 'Minas Gerais'],
    'comprou': [True, False, True]
}
```

```python
# criando um DataFrame
df = pd.DataFrame(dados)
```

```python
df
```

```python
#criando id do usuário
id_usuario = [867, 567, 918]
```

```python
# associar o id do usuário a lista
df.index = id_usuario
```

```python
df
```

```python
dados = [['Marcos', 29, 'São Paulo', 'True'],
         ['Miguel', 59, 'Bahia', 'False'],
         ['Angelina', 60, 'Minas Gerais', 'True' ]]

# Criando dataframe
df = pd.DataFrame(dados, columns=['nome', 'idade', 'cidade', 'comprou'],
                  index=[867, 567, 918])
```

```python
#visualizar
df
```

```python
#utilizando a função iloc para acessar linhas
df.loc[867]
```

```python
#criando novas colunas
df['saldo'] = 0.0

df
```

```python
#alterando index do df
df.index = ['a', 'b', 'c']
df
```

```python
# alterar os labels das colunas
df.columns = ['Nome do Cliente', 'Idade', 'Naturalidade', 'Fez Compras?', 'Saldo na Loja']

df
```
